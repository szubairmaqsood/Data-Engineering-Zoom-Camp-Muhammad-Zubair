```python
import pandas as pd


df = pd.read_parquet("green_tripdata_2025-11.parquet")


print(df.shape)
print(df.head())
```

```python
df.info()
```

```python
count = (df["trip_distance"] <= 1).sum()
count
```

```python
df.isna().sum()
```

```python
nov_df = df[
(df["lpep_pickup_datetime "] >= "2025-11-01") &
(df["lpep_pickup_datetime"] < "2025-12-01")
]
```

```python
count = (nov_df["trip_distance"] <= 1).sum()
count
```

```python
nov_df1 = df[(df["trip_distance"] < 100)]
daily_max = (
nov_df1
.groupby("lpep_pickup_datetime")["trip_distance"]
.max()
)
daily_max.idxmax(), daily_max.max()
```

```python
# 2) filter trips on Nov 18, 2025 (date-only filter; works when column is already datetime)
_18_nov_df = df[df["lpep_pickup_datetime"].dt.date == pd.Timestamp("2025-11-18").date()]


# 3) sum total_amount by pickup location
zone_totals = _18_nov_df.groupby("PULocationID")["total_amount"].sum()


# 4) pickup zone (ID) with the largest total_amount + the amount
zone_totals.idxmax(), zone_totals.max()
```
